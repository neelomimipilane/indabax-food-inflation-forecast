Exploratory Analysis



```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))

from src.preprocessing import load_data, add_features, split_train_test
from src.models import train_sarimax, train_lstm
from src.evaluation import evaluate_models

data = load_data()
features = add_features(data)
train_data, test_data = split_train_test(features)
print(f"Data shape: {features.shape}")
print(f"Train: {train_data.shape}, Test: {test_data.shape}")
features.head()
```

```python
import matplotlib.pyplot as plt

# quick look at inflation
plt.figure(figsize=(10, 3))
plt.plot(features.index, features['food_price_inflation'])
plt.title('Food Price Inflation')
plt.show()
```

```python
# Run SARIMAX
sarimax_result = train_sarimax(train_data, test_data)
print('SARIMAX order:', sarimax_result['order'])
print('SARIMAX AIC:', sarimax_result.get('aic'))
for k, v in sarimax_result['metrics'].items():
    print(f'{k}: {v:.3f}')
```

```python
# Run LSTM
lstm_result = train_lstm(train_data, test_data)
if lstm_result is not None:
    for k, v in lstm_result['metrics'].items():
        print(f'{k}: {v:.3f}')
else:
    print('LSTM skipped: TensorFlow not installed')
```

```python
# Generate all outputs
from pathlib import Path
root = Path.cwd().parent
evaluate_models(
    sarimax_result,
    lstm_result,
    features,
    test_data,
    root / 'figures',
    root / 'outputs',
)
```
